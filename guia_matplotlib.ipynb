Matplotlib es una librería de Python que permite realizar gráficas 2D de excelente calidad. Es multiplataforma y puede ser usada desde scripts o desde la consola de Python.  

>Podemos definir el objetivo de matplotlib con las propias palabras de su creador John Hunter -- "matplotlib tries to make easy things easy and hard things possible" --  


Matplotlib fue creado tomando como base a Matlab por lo que para las personas que han gráficado alguna vez en su vida con Matlab, Matplotlib les va a resultar muy familiar. Podemos exportar nuestras gráficas en los formatos de imágenes mas populares e incluso a formato latex para su inclusión en artículos científicos.  


**¿Si Matplotlib fue creado pensando en Matlab, porqué debería usarlo en vez de usar el mismo Matlab?**    


Existen varias razones. Es cierto que Matplotlib se baso en Matlab pero específicamente en su facilidad de uso. Matplotlib no trata de ser un reemplazo a Matlab como si lo es Octave. Matplotlib es un librería para Python que permite realizar gráficas brindando múltiples facilidades para lograr ese objetivo. Si queremos realizar gráficas de cualquier tipo y no necesitamos toda la cantidad de toolboxes que vienen con Matlab, matplotlib es justo lo que necesitamos.
Por otro lado, y no menos importante. Matplotlib es una librería libre lo cual quiere decir entre otras cosas que podemos usarla completamente gratis, lo cual es una ventaja de entrada frente a Matlab.

### Leneg-label-axis

````python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
````

````python
#plotear todas las variables
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
````

##### Agregar titulo y nombre a los ejes x y

````python
#titulo
plt.title('el clima')

#Nombre de ejes x,y
plt.xlabel('Días de la semana')
plt.ylabel('Temperatura')

plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
````

##### Agregar leyenda y mostrarla

````python
#label = 'nombre de la leyenda'
plt.plot(days, max_t, label = 'max')
plt.plot(days, min_t, label = 'min')
plt.plot(days, avg_t, label = 'avg')

#para mostrar la leyenda, por defecto la posiciona en el mejor lugar si queremos lugar especifico
#escribiomos dentro ejemplo : loc = 'upper left' 
plt.legend()
````

````python
#Tamaño del grafico
plt.figure(figsize = (12,5))

#grafica
plt.plot(days, max_t, label = 'max')
plt.plot(days, min_t, label = 'min')
plt.plot(days, avg_t, label = 'avg')

#ver leyendas y agregarle formato
plt.legend(loc = 'upper right', shadow = True,fontsize = 'large')
````

##### Grid 

````python
plt.figure(figsize = (12,5))
plt.plot(days, max_t, label = 'max')
plt.plot(days, min_t, label = 'min')
plt.plot(days, avg_t, label = 'avg')
plt.legend(loc = 'upper right', shadow = True,fontsize = 'large')

#grid
plt.grid()
````

## Subplots

````python
# agregar estilo al grafico
plt.style.use('seaborn')

# graficar dos graficas separadas
fig, (ax1, ax2) = plt.subplots(2,1, sharex=True)

#plt.figure(figsize = (12,5))
ax1.plot(days, max_t, label = 'max')
ax1.plot(days, min_t, label = 'min')
ax2.plot(days, avg_t, label = 'avg')

#titulo 
ax1.set_title('Gráficas del clima',size = 14)
ax1.set_ylabel('Temepratura')
ax1.legend(loc = 'upper right', shadow = True,fontsize = 'large')

ax2.set_ylabel('Temepratura')
ax2.set_xlabel('Días de la semana', size = 10)
ax2.legend(loc = 'upper right', shadow = True,fontsize = 'large')

plt.tight_layout()
plt.show()
````

## Barchart

##### Gráfico de barra simple

````python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
````

### Gráfico de barras con sus componentes

````python
plt.title('ventas company tech')
#colocar etiquetas al eje x
plt.xticks(xpos, company)


plt.bar(xpos, revenue, label ='revenue')
plt.ylabel('Ventas en Billones USD$')
plt.legend()
````

##### Gráfica de barras stack

````python
profit=[40,2,34,12]
````

````python
plt.bar(xpos, revenue, label = 'revenue')
plt.bar(xpos, profit, label = 'profit')

plt.title('ventas company tech')
plt.xticks(xpos, company)
plt.ylabel('Ventas en Billones USD$')
plt.legend()
````

##### Barras Multiples

````python
#agregarle -0.2 y +0.2 tambien, width = 0 a 1 el ancho de las barras
plt.bar(xpos-0.2, revenue, label = 'revenue', width = 0.4)
plt.bar(xpos+0.2, profit, label = 'profit', width = 0.4)

plt.title('ventas company tech')
plt.xticks(xpos, company)
plt.ylabel('Ventas en Billones USD$')
plt.legend()
````

##### Barras horizontales

````python
plt.title('ventas company tech')
#colocar etiquetas el eje y
plt.yticks(xpos, company)


plt.barh(xpos, revenue, label ='revenue')
#colocar subtitulo al eje x
plt.xlabel('Ventas en Billones USD$')
plt.legend()
````

### Hitograma  
In histograms, x axis contains a variable and y axis will be a frequency of that variable

````python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
````

````python
plt.hist(blood_sugar)
````

##### separar las barras del histograma

````python
#el parametro rwidth separa las barras en un histograma
plt.hist(blood_sugar, rwidth=0.8)
````

````python
# otra forma de separación, o ver las barras marcadas edgecolor y linewidth

plt.hist(blood_sugar, edgecolor = 'black', linewidth=2)
plt.show()
````

### parametro bins

````python
#bins de cuatro bloques o grupos
plt.hist(blood_sugar, bins=4, rwidth=0.5)
````

##### Histogram con clasifiación de grupos  

- 80-100: Normal
- 100-125: Pre-diabetic
- 125-150: Diabetic

````python
plt.xlabel('Nivel de azucar')
plt.ylabel('cantidad de pacientes')
plt.title('Gráfica de niveles de azucar en la sangre')
plt.hist(blood_sugar, bins= [80, 100, 125, 150], rwidth=0.65, color='g')
plt.show()
````

##### Multiple datos

````python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
````

````python
plt.figure(figsize=(10,4.5))
plt.xlabel('Nivel de azucar', size = 12)
plt.ylabel('cantidad de pacientes', size=12)
plt.title('Gráfica de niveles de azucar en la sangre', size = 16)

plt.hist([blood_sugar_men, blood_sugar_women], bins =[80, 100, 125, 150], rwidth=0.95, color=['green', 'orange'], 
         label =['men', 'women'])

plt.legend(shadow= True, fontsize = 'large')
plt.show()
````

### Histtype = step

````python
plt.figure(figsize=(10,4.5))
plt.xlabel('Nivel de azucar', size = 12)
plt.ylabel('cantidad de pacientes', size=12)
plt.title('Gráfica de niveles de azucar en la sangre', size = 16)

plt.hist(blood_sugar, bins = [80,100,125,150], rwidth=0.95, histtype='step')
plt.show()
````

##### Cambiar a horizontal

````python
plt.figure(figsize=(10,4.5))
plt.ylabel('Nivel de azucar', size = 12)
plt.xlabel('cantidad de pacientes', size=12)
plt.title('Gráfica de niveles de azucar en la sangre', size = 16)

plt.hist(blood_sugar, bins= [80,100,125,150], rwidth=0.95, orientation='horizontal')

plt.show()
````

## Pie chart

````python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
````

````python
# simple pie_chart
plt.pie(exp_vals, labels = exp_labels)
plt.show()
````

#####  circulo perecto

````python
plt.pie(exp_vals, labels = exp_labels)
plt.axis('equal')
plt.show()
````

##### Agregarle el procentaje y radio

````python
#porciento y un radio mas grande
plt.pie(exp_vals, labels = exp_labels,autopct='%1.2f%%', shadow = True, radius=1.5)
plt.axis('equal')
plt.show()
````

##### Resaltar o explode 

````python
plt.style.use('ggplot')
plt.axis('equal')
plt.pie(exp_vals, labels = exp_labels,autopct='%1.2f%%', shadow = True, radius=1.5, explode=[0,0.2,0,0,0])
plt.show()
````

##### Angulo y counterclok

````python
plt.style.use('fivethirtyeight')
plt.axis('equal')
plt.pie(exp_vals, labels = exp_labels,autopct='%1.2f%%', shadow = True, radius=1.5, explode=[0,0.2,0,0,0.2], startangle=45,
       counterclock=True)
plt.show()
````

## stack plot

````python
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]
player1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]
````

````python
#simple stackplot
plt.stackplot(minutes, player1, player2, player3)
plt.show()
````

````python
# agregarle etiquetas, titulo y leyenda-color
labels = ['player1', 'player2', 'player3']
colors = ['#6d904f', '#fc4f30', '#008fd5']

plt.stackplot(minutes, player1, player2, player3, labels=labels, colors=colors)
plt.title('Gráfica de tiempo de juego')
plt.tight_layout()
plt.legend(loc='upper left', shadow=True)
plt.show()
````

## Gráficar live data

````python
import random
from itertools import count
from matplotlib.animation import FuncAnimation
````

````python
# creare una funcion que se asemeje a los datos en linea
x_vals =[0, 1, 2, 3, 4, 5]
y_vals =[0, 1, 2, 3, 4, 5]
index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0,5))
    plt.cla()
    plt.plot(x_vals, y_vals)    
    plt.tight_layout()
    
ani = FuncAnimation(plt.gcf(),animate, interval=1000)
plt.show()
````

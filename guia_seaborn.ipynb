---

**Seaborn** es una envoltura alrededor de Matplotlib que facilita la creación de gráficos estadísticos comunes. La lista de gráficos admitidos incluye gráficos de distribución univariada y bivariada, gráficos de regresión y varios métodos para representar variables categóricas. La lista completa de las parcelas que proporciona Seaborn se encuentra en su referencia API .  

Crear gráficos en Seaborn es tan simple como llamar a la función de gráficos apropiada. Este es un ejemplo de la creación de un histograma, una estimación de la densidad del kernel y una gráfica de tap para datos generados aleatoriamente.  

---

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

### Como importar seaborn

````python
import seaborn as sns
````

### Carga del dataset de ejemplo  tips

````python
tips = sns.load_dataset('tips')
tips.head()
````

### Relational plot 

Este gráfico utiliza como parametros obligatorios los  ejes: 
+ x = columna1
+ y= columna2
+ data= el archivo  

Es utilizado con datos de tipo númericos los parametros que utiliza se le puede colocar las columnas con datos catégoricos

````python
sns.relplot(x= 'total_bill',y = 'tip', data= tips )
````

````python
# tambien podemos realizar el grafico de esta forma sin especificar ya que tiene un orden especifico
sns.relplot('total_bill','tip', data = tips)
````

### Parametro Hue  
este parametro puede utilizarse con las columnas categorícas

````python
sns.relplot(x='total_bill', y = 'tip', data = tips, hue= 'sex' )
````

````python
sns.relplot(x='total_bill', y = 'tip', data = tips, hue= 'smoker' )
````

### parametro style 

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', style='smoker' )
````

### Parametro size

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size ='sex' )
````

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size= 'smoker' )
````

### Line Chart  

Utiliza los siguientes parametros principales:
- x = nombre_columna
- y = nombre_columna
- data = nombre del archivo
- kin = 'line

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, kind='line')
````

#### Parametro ci = indeice de confianza

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, kind = 'line', ci = 'sd')
````

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, kind = 'line', style='smoker')
````

### Parametro col

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, kind='line',
           col = 'smoker')
````

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, kind='line',
           col = 'day')
````

#### Parametro col_wrap

````python
sns.relplot('total_bill', 'tip', data = tips, hue = 'sex', kind='line', 
            col='day', col_wrap=2)
````

### Catplot  

para comparar los datos de columnas categorícas, se le coloca la columna númerica para saber su dimensionalidad

````python
sns.catplot('sex', 'total_bill', data = tips)
````

````python
sns.catplot('day', 'tip', data = tips)
````

#### Parametro Jitter  
tine que ver con la magnito de los datos o la anchura de estas barras

````python
sns.catplot('day', 'total_bill', data = tips, jitter = False)
````

````python
sns.catplot(x = 'day', y = 'total_bill',  data = tips.query('total_bill <30'), jitter = False)
````

````python
sns.catplot(x = 'day', y = 'total_bill',  data = tips.query('tip > 5'), jitter = False)
````

#### Parametro kin='swarm'

````python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm')
````

#### Tips para cambiar el grafico a horizontal

solo debes de cambiar los ajes x y:

````python
sns.catplot(y = 'day', x = 'total_bill', data = tips, kind = 'swarm')
````

````python
sns.catplot(y ='day', x ='tip', data = tips)
````

#### Parametro Row: 
Es lo contrario del parametro col

````python
sns.catplot('sex', 'tip', data = tips, row= 'day')
````

### BOX plot

````python
sns.catplot('day', 'tip', data = tips, kind = 'box')
````

````python
sns.catplot('smoker', 'total_bill', data = tips, kind='box',
           hue= 'day')
````

#### Parametro hue_order  
nos permite rearle un orden propio

````python
sns.catplot('smoker', 'total_bill', data = tips, kind='box',
           hue= 'day', hue_order=['Sun', 'Thur', 'Fri', 'Sat'])
````

````python
sns.catplot('smoker', 'total_bill', data = tips, kind='box', hue='sex'
           ,col='day', col_wrap=2)
````

````python
sns.catplot(y= 'day', x= 'tip', data = tips, kind = 'box')
````

### Violin Plot 

````python
sns.catplot('day', 'total_bill', data = tips, kind='violin')
````

#### Parametro inner

````python
sns.catplot('day', 'total_bill', data = tips, kind='violin', inner = 'stick')
````

### Bar chart

````python
sns.catplot('day', 'total_bill', data = tips, kind='bar')
````

### Point Plot

````python
sns.catplot('day', 'total_bill', data = tips, kind = 'point')
````

````python
sns.catplot('day', 'total_bill', data = tips, kind = 'point', hue= 'sex')
````

````python
sns.catplot('sex', 'total_bill', data = tips, kind = 'point', hue = 'smoker',col= 'day', col_wrap=2)
````

### Displot 
Gráfica de distribución de los datos

````python
sns.distplot(tips.total_bill)
````

#### Parametro kde  
elimina la linea de distribución

````python
sns.distplot(tips.total_bill, kde= False)
````

#### parametro hits  
elimina las barras o histogramas

````python
sns.distplot(tips.total_bill, hist=False)
````

#### parametro rug  

indica donde hay mas concentración de los datos

````python
sns.distplot(tips.total_bill, rug=True)
````

````python
sns.distplot(tips.total_bill[tips.sex == 'Female'], rug=True )
````

````python
sns.distplot(tips.total_bill[(tips.sex == 'Female') & (tips.smoker=='No')], rug=True )
````

````python
sns.kdeplot(tips.total_bill, bw=0.2)
````

### Joinplot
es una especie de scatter plot, pero en las esquinas nos da como estan distribuidos los datos

````python
sns.jointplot('total_bill', 'tip', data = tips)
````

````python
sns.jointplot('total_bill', 'size', data = tips, color = 'red')
````

#### parametro kind
- kind = 'hex'
- kind ='kde'

````python
sns.jointplot('total_bill', 'size', data = tips, color = 'red', kind ='hex')
````

````python
sns.jointplot('total_bill', 'size', data = tips, color = 'red', kind ='kde')
````

### Heatmap

````python
sns.heatmap(tips.corr())
````

#### Parametro  
- annont = True coloca los numero dentro de cada casilla

````python
sns.heatmap(tips.corr(), annot=True)
````

- fmt = le da formato a los numeros

````python
sns.heatmap(tips.corr(), annot=True, fmt='.0%')
````

- cmap = palestra de color

````python
sns.heatmap(tips.corr(), annot=True, fmt='.0%', cmap= 'BuPu')
````

````python
plt.style.use('ggplot')
sns.heatmap(tips.corr(), annot=True, fmt='.0%', cmap='viridis')
````

- linewidths resalta las lineas

````python
plt.style.use('ggplot')
sns.heatmap(tips.corr(), annot=True, fmt='.0%', cmap='viridis', linewidths=2)
````
